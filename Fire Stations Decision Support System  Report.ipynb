I download data in to my computer as an csv file. Let's look into our data.

```python
import pandas as pd 

df = pd.read_csv('FDNY_Firehouse_Listing.csv')
df
```

We remove the columns which we will not use in our analysis.

```python
df=df.drop(['Community Board', 'Community Council ', 'Census Tract',
       'BIN', 'BBL', 'NTA'],axis=1)
```

Now we delete the empty rows.

```python
df = df.dropna()
```

```python
import folium
```

In order to visualize our code we use folium. First we draw circle to center of the new york.

```python
longitude=-74.0153365
latitude=40.7149555
lmap = folium.Map(location=[latitude, longitude], zoom_start=12) 
folium.CircleMarker(
    [latitude, longitude],
    radius=12,
    color='red',
    popup='NYC',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(lmap)

lmap
```

Now we are going to show the locations of the fire stations.

```python
for lat, lng, in zip(df.Latitude, df.Longitude):
    lmap.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=10, # define how big you want the circle markers to be
            color='red',
            fill=True,
            fill_color='blue',
            fill_opacity=0.8
        )
    )
lmap
```

# Methodology


In this project we will direct our efforts on detecting areas of NYC that have low fire station density, particularly those with low number of fire stations.

In first step we have collected the required data: location and type (category) of every fire station in NYC  We have also identified fire stationss according to  https://data.cityofnewyork.us/widgets/hc8x-tcnd data.


In second step we classified our data according to the boroughs. 

In third step we analyze the descriptive statistics of the fire station numbers.
