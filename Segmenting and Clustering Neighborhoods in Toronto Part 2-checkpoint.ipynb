```python
import pandas as pd 
cdf=pd.read_csv('Geospatial_Coordinates.csv')
cdf=cdf.rename(columns={"Postal Code": "Postcode"})
cdf.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
table=soup.find('table', class_='wikitable sortable')
rows = table.find_all('tr')
columns = [v.text.replace('\n','') for v in rows[0].find_all('th') ]
```

```python
df = pd.DataFrame(columns=columns)

for i in range(1, 265):
    tds = rows[i].find_all('td')
    if len(tds) == 3:
   
        values = [tds[0].text.replace('\n',''),tds[1].text.replace('\n',''),tds[2].text.replace('\n','')]
    
        if values[1]=='Not assigned':
            values=[]
        elif values[1]!='Not assigned' and values[2]=='Not assigned':
            values[2]=values[1]
            df = df.append(pd.Series(values, index=columns),ignore_index=True)

        else:
            
            df = df.append(pd.Series(values, index=columns),ignore_index=True)
 
    else:
        
        values = [td.text for td in tds]
        
   
print(df)
```
